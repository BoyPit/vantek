```python
from langchain_community.document_loaders import ApifyDatasetLoader

# Define a mapping function to extract relevant data
def mapping_function(item):
    return item  # Modify if necessary to structure your data properly

# Load dataset with the required mapping function
loader = ApifyDatasetLoader(
    dataset_id="1pDGp8zdGxyNSz4SI",
    dataset_mapping_function=mapping_function
)

# Load the dataset
grouped_documents = loader.load()

print(grouped_documents)
# Extract URLs from organicResults
def extract_links(data):
    links = []
    for item in data:
        if "organicResults" in item:
            for result in item["organicResults"]:
                if "url" in result:
                    links.append(result["url"])
    return links

# Get all URLs
all_links = extract_links(grouped_documents)

# Print extracted links
print(all_links)
i=0
for link in all_links:
    print(link)
```
